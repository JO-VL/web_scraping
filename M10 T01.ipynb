#### BeautifulSoup quotes

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url='http://quotes.toscrape.com/'
doc=BeautifulSoup(requests.get(url).text,'html.parser')
print(doc)
```

Examinem el codi font i veiem que les cites són de classe "text", els autors són de classe "author" i els tags són de classe "tags".

```python
# Cites
cites=[]
div_tags=doc.find_all('div',class_='quote')
for tag in div_tags:
    cita=tag.find('span', class_ ='text').text
    cites.append(cita)

cites
```

```python
# Autors de les cites
autors=[]
for tag in div_tags:
    span_tag=tag.find('span', class_=None)
    autor=span_tag.find('small', class_='author').text
    autors.append(autor)

autors
```

```python
# Tags de les cites
tags_cites=[]
for tag in div_tags:
    tag_cita=tag.find('div',class_='tags').meta['content']
    tags_cites.append(tag_cita)
    
tags_cites
```

```python
# Fem un dataframe amb la informació extreta
df = pd.DataFrame()
df['Cita']= cites
df['Autors']=autors
df['Tags']=tags_cites
df
```

#### Selenium quotes

```python
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By

driver = Chrome()

driver.get("https://quotes.toscrape.com")
```

```python
cites2 = driver.find_elements(By.CLASS_NAME, "text")

for cita in cites2:
    cita = cita.text.strip()
    print(cita)
```

```python
autors2 = driver.find_elements(By.CLASS_NAME, "author")

for autor in autors2:
    autor= autor.text.strip()
    print(autor)
```

```python
tags = driver.find_elements(By.CLASS_NAME, "tags")

for tag in tags:
    tag= tag.text.strip()
    print(tag)
```

```python
# Fem un dataframe amb la informació extreta

cites2_text = [cita.text.strip() for cita in cites2]
autors2_text = [autor.text.strip() for autor in autors2]
tags_text = [tag.text.strip() for tag in tags]

df2 = pd.DataFrame()
df2['Cita']= cites2_text
df2['Autors']=autors2_text
df2['Tags']=tags_text
df2['Tags'] = df2['Tags'].str.replace('Tags: ', '')
df2
```

```python
# Imprimim les categories

URL2 = "http://books.toscrape.com/index.html"
page = requests.get(URL2)

soup = BeautifulSoup(page.content, 'html.parser')

categories = []
side_categories = soup.find('div', class_='side_categories')
a_tags = side_categories.find_all('a')
for tag in a_tags:
    categoria = tag.text.strip()
    categories.append(categoria)

print(categories)
```
